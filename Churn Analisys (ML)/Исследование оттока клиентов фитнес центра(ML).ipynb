### Изучим распределение по возрасту

```python
def calculate_percentage(row1, row2):
    return (row1 / row2) * 100
```

Функция для определения процента от суммы

```python
ages_churn = churn['age'].value_counts().reset_index()
ages_churn.columns = ['age', 'count']
ages_churn['sum'] = ages_churn['count'].sum()
ages_churn['percentile'] = calculate_percentage(ages_churn['count'], ages_churn['sum'])
```

```python
ages_clients = clients['age'].value_counts().reset_index()
ages_clients.columns = ['age', 'count']
ages_clients['sum'] = ages_clients['count'].sum()
ages_clients['percentile'] = calculate_percentage(ages_clients['count'], ages_clients['sum'])
```

Сводные таблицы по возрасту

```python
plt.bar(x=ages_churn['age'], height = ages_churn['percentile'], label='Отток', alpha=0.7)
plt.bar(x=ages_clients['age'], height = ages_clients['percentile'], label='Нет', alpha=0.7)

plt.legend(title='Группа оттока')
plt.title('Распределение клиентов по возрасту и группе оттока')
plt.xlabel('Возраст')
plt.ylabel('Процент от всех клиентов в группе');
```

Видим что график для "отвалившихся" смещен влево, что говорит о том, что группа оттока немного моложе

## Построим модель прогнозирования оттока клиентов

Разделим выборку на целевую переменную и признаки

```python
X = clean_data.drop(['churn'], axis=1)
y = clean_data['churn']
```

Затем разделим выборку на обучающую и валидационную

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Стандартизируем признаки

```python
scaler = StandardScaler()

X_st_train = scaler.fit_transform(X_train)
X_st_test = scaler.transform(X_test)
```

### Обучим модель алгоритмом LogisticRegression

```python
model_lg = LogisticRegression(solver='liblinear', random_state=0)
model_lg.fit(X_st_train, y_train)
```

Получим предсказания

```python
y_pred_lg = model_lg.predict(X_st_test)
```

Узнаем метрики классификации для модели Логистической регрессии

```python
print('Для модели Логистической Регрессии метрики классификации следующие:')
print(f'Доля правильных ответов (accuracy score) равна: ', accuracy_score(y_test, y_pred_lg).round(4) * 100,'%')
print(f'Точность (precision score) равна: ', precision_score(y_test, y_pred_lg).round(4) * 100, '%')
print(f'Полнота (recall) равна:', recall_score(y_test,y_pred_lg).round(3) * 100, '%')
```

### Обучим модель RandomForestClassifier

Для этого повторим те же шаги

```python
model_rf = RandomForestClassifier(random_state=0)
model_rf.fit(X_st_train, y_train)
```

```python
y_pred_rf = model_rf.predict(X_st_test)
```

```python
print('Для модели Случайного Леса метрики классификации следующие:')
print(f'Доля правильных ответов (accuracy score) равна: ', accuracy_score(y_test, y_pred_rf).round(4) * 100,'%')
print(f'Точность (precision score) равна: ', precision_score(y_test, y_pred_rf).round(3) * 100, '%')
print(f'Полнота (recall) равна:', recall_score(y_test,y_pred_rf).round(4) * 100, '%')
```

Выводы : 

В целом обе модели ведут себя неплохо, показывают примерно одинаковую долю правильных ответов на валидационной выборке, но так как классы в нашей выборке несбалансированы, смотрим на значения точности и полноты. Точность чуть выше у второй модели (на 2%), тогда как значение полноты меньше на 7%, так что если выбирать более удачную модель, я бы отдал предпочтение LogisticRegression из-за чуть лучшего баланса.
