## Шаг 4 Проверка статистических гипотез:

### Проверим гипотезу о различии возраста между клиентами, пользующимися двумя продуктами банка, и теми, кто пользуется одним

_$H_0$ - Возраст между клиентами с двумя продуктами равен возрасту клиентов с одним_

_$H_1$ - Возраст в двух группах различен_

Посмотрим на распределения для изучаемых групп

```python
sns.distplot(data[data['products'] == 1]['age'], color='blue', label=1, hist_kws={'edgecolor':'black'})
sns.distplot(data[data['products'] == 2]['age'], color='red', label=2, hist_kws={'edgecolor':'black'})

plt.title('Распредление плотности по возрасту')
plt.ylabel('Плотность вероятности')
plt.xlabel('Возраст')
plt.legend(title='Количество продуктов');
```

Видим в обоих случаях распределения, близкие к нормальным. Так же по линии плотности можно судить о схожести дисперсий

```python
print('\033[1m' + 'Количество наблюдений по группам:' + '\033[0m')
data.query('products == [1,2]').groupby('products')['id'].count()
```

Количество наблюдений велико, распределено примерно 1:1.5

Так как мы имеем нормальное распределение выборочных средних (согласно ЦПТ), количество наблюдений велико и распределено достаточно равномерно (1:1.5), наши выборки независимы друг от друга, в данных о возрасте нет выбросов ([что было изучено ранее](#Распределение-клиентов-по-возрасту)), для проверки гипотезы будем использовать t-критерий.

В данном случае гипотеза на равенство средних, значит будем проверять её t-тестом __t_test_ind__

```python
a = 0.05

print(f'Альфа равно : ', a*100, '%')

result = st.ttest_ind(data[data['products'] == 2]['age'], data[data['products'] == 1]['age'])

print(f'P-value равно : ', result.pvalue * 100,'%')
```

Вывод :

При уровне статистической значимости в 5% процентов у нас есть основания отвергнуть нулевую гипотезу в пользу альтернативной, так как P-value равно 3.35%. 

Гипотеза о равенстве возрастов между группами отвергнута, гипотеза о различии возрастов - не отвергнута, так как мы видим статистически значимые различия между группами

#### Проверим гипотезу о различии кредитного рейтинга у пользователей одного продукта и двух

_$H_0$ - Рейтинг клиентов с двумя продуктами равен рейтингу клиентов с одним_

_$H_1$ - Рейтинг первой группы выше рейтинга второй_

```python
sns.histplot(data.query('products == [1,2]'), x='score', hue='products', kde=True, common_norm=False, stat='density', 
             palette=['blue', 'red'])

plt.title('Распредление плотности по рейтингу')
plt.ylabel('Плотность вероятности')
plt.xlabel('Рейтинг');
```

```python
print('\033[1m' + 'Описательная статистика по группам:' + '\033[0m')
data.query('products == [1,2]').groupby('products').describe()['score']
```

Так как мы имеем нормальное распределение выборочных средних (согласно ЦПТ), количество наблюдений велико и распределено достаточно равномерно (1:1.5), в данных о рейтинге нет выбросов ([что было изучено ранее](#Распределение-по-кредитному-рейтингу)), для проверки гипотезы будем использовать t-критерий, как и в предыдущем случае, с той лишь разницей, что здесь мы можем предпологать разные дисперсии в выборках

В данном случае гипотеза на равенство средних, значит будем проверять её t-тестом __t_test_ind__, так же зададим парамерт _equal_var=False_, так как есть сомнения в равенстве дисперсий. Так же учитывая, что альтернативная гипотеза у нас правосторонняя, зададим параметр _alternative='greater'_

```python
a = 0.05

print(f'Альфа равно : ', a*100, '%')

result = st.ttest_ind(data[data['products'] == 1]['score'], data[data['products'] == 2]['score'],
    alternative='greater', equal_var=False)

print(f'P-value равно : ', result.pvalue * 100,'%')
```

Вывод :

При уровне статистической значимости в 5% процентов у нас есть основания отвергнуть нулевую гипотезу в пользу альтернативной, так как значение P-value крайне низкое (менее 0.1 процента)

Гипотеза о равенстве рейтингов между группами отвергнута, гипотеза о том, что рейтинг клиентов, пользующихся одним продуктом, выше рейтинга клиентов, пользующихся двумя - не отвергнута, так как мы видим статистически значимые различия между группами
