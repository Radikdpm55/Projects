## Проект "Исследование компьютерных игр"

### Описание проекта

<font size=3> Имеются данные из открытых источников о продажах компьютерных игр до 2016 года, нужно выявить определяющие успешность игры закономерности, что позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании на 2017 год интернет-магазина игр "Стримчик"

### Описание данных

- _Name_ — название игры
- _Platform_ — платформа
- _Year_of_Release_ — год выпуска
- _Genre_ — жанр игры
- _NA_sales_ — продажи в Северной Америке (миллионы проданных копий)
- _EU_sales_ — продажи в Европе (миллионы проданных копий)
- _JP_sales_ — продажи в Японии (миллионы проданных копий)
- _Other_sales_ — продажи в других странах (миллионы проданных копий)
- _Critic_Score_ — оценка критиков (максимум 100)
- _User_Score_ — оценка пользователей (максимум 10)
- _Rating_ — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.


Данные за 2016 год могут быть неполными.

### Посмотрим на распределение игр по рейтингу

```python
cm = plt.cm.get_cmap('viridis_r')
colors = [cm(v) for v in np.linspace(0,1,5)]

data.query('rating != "unknown"').value_counts('rating')[0:4].plot(kind='pie',
                            colors=colors, autopct='%.1f%%', explode=[0.1,0,0,0])
plt.title('Распределение игр по рейтингу ESRB')
plt.ylabel('');
```

У нас есть данные о 4 наиболее представленных категориях, расшифруем их значение; согласно данным из [википедии](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board):
- E («Everyone») — «Для всех»:  Содержание вполне подходит для всех возрастов (ранее с 6 лет)
- T («Teen») — «Подросткам» : Игра подходит для лиц от 13 лет
- M («Mature») — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет
- E10+ («Everyone 10 and older») — «Для всех от 10 лет и старше»

Соответственно видим из диаграммы что больше всего игр категории E(для всех) - более 40%. Чуть меньше 30% игр представлено из категории T(подросткам). И по ~15% игр категорий M(для взрослых) и E10+(Для всех от 10 лет и старше)

### Изучим количество игр по годам

```python
years = data.groupby('year', as_index=False)[['name']].count().sort_values(by='year', ascending=False)
years.columns = ['year', 'games_cnt']

plt.figure(figsize=[8,5])

sns.lineplot(data=years, x='year', y='games_cnt')

plt.xticks(range(1980,2017,3))
plt.axhline(y=100, color='grey', alpha=0.4, linestyle='--') 
plt.title('Количество выпускаемых игр по годам')
plt.xlabel('Год выпуска')
plt.ylabel('Количество выпущенных игр за год');
```

Видим, что до 1993 года количество выпускаемых игр не превышало 100, далее к 1998 году оно увеличивается до 400, затем еще один резкий скачок в 2000-2002 годах, даллее с 2004 по 2008-2009 годы, где достигает своего пика в количестве более 1400 игр за год. После 2010 года выпуск резко падает к показателю около 500 игр за 2016 год. Если говорит об актуальном периоде, то можно вынести за скобки период с 1980 по 1993 годы, так как в это время индустрия только зарождалась

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

$H_0$ - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

$H_1$ - Средние пользовательские рейтинги платформ Xbox One и PC отличаются

Посмотрим на корректность сравниваемых данных

```python
test_data = actual_data.query('user_score != 123').dropna(subset=['user_score'])
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=test_data.query('platform in ["PC", "XOne"]'), y='platform', x='user_score')

plt.title('Распределения пользовательских оценок для платформ PC и Xbox One', pad=10)
plt.ylabel('Платформа')
plt.xlabel('Оценка');
```

```python
print('Количество записей о каждой платформе')
test_data.query('platform in ["PC", "XOne"]').groupby('platform').agg({'name':'count'})
```

Как мы видели [ранее](#Распределения-данных), мы имеем нормальное распределение оценок с небольшим хвостом влево. Выбивающиеся значения лежат не так далеко от центра данных, и их не большое количество, так что не вижу смысла их удалять, оставим

Так как у нас гипотезы на равенство средних и выборки независимы, а так же количество наблюдений велико, проверим их T-тестом *ttest_ind*, уровень альфа зададим в 5%

```python
a = 0.05

print(f'Альфа равно : ', a*100, '%')

result = st.ttest_ind(test_data[test_data['platform'] == 'XOne']['user_score'], 
                                         test_data[test_data['platform'] == 'PC']['user_score'])

print(f'P-value равно : ', round(result.pvalue * 100, 2),'%')
```

Видим значение p-value гораздо выше alpha, что не даёт нам основания отклонить нулевую гипотезу - cредние пользовательские рейтинги платформ Xbox One и PC не имеют статистически значимых различий в актуальный период.

### Средние пользовательские рейтинги жанров Action и Sports разные

$H_0$ - Средние пользовательские рейтинги жанров Action и Sports одинаковые

$H_1$ - Средние пользовательские рейтинги жанров Action и Sports разные

Так же посмотрим на распределение оценок

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=test_data.query('genre in ["Action", "Sports"]'), y='genre', x='user_score')

plt.title('Распределение пользовательских оценок для жанров Action и Sports', pad=11)
plt.ylabel('Жанр')
plt.xlabel('Оценка');
```

```python
print('Количество оценок о каждом жанре')
test_data.query('genre in ["Action", "Sports"]').groupby('genre').agg({'name':'count'})
```

Здесь чуть больше выбивающихся оценок ниже 4 в жанре Action, которые так же недалеко от центра данных, оставим, учитывая, что и наблюдений у нас в этом случае больше.

Проверим гипотезы аналогично предыдущему разу

```python
a = 0.05

print(f'Альфа равно : ', a*100, '%')

result = st.ttest_ind(test_data[test_data['genre'] == 'Action']['user_score'], 
                                         test_data[test_data['genre'] == 'Sports']['user_score'])

print(f'P-value равно : ', (result.pvalue * 100),'%')
```

Крайне низкое значение p-value даёт повод нам отвергнуть нулевую гипотезу, которая звучит как "Средние пользовательские рейтинги жанров Action и Sports одинаковые", в пользу альтернативной - рейтинги различаются

Результаты теста говорят о том, что мы видим статистически значимые различия в пользовательских оценках между жанром Action и Sports

Так как нулевая гипотеза должна быть сформулирована со знаком равенства, мы приняли данное утверждение за нулевую гипотезу. Альтернативную сформуливали со знаком неравенства.
